So far we've been training and evaluating models with default values for hyperparameters.

Today we will perform the same feature engineering as before, and then compare the best working models you got so far, but now fine tuning it's hyperparameters.

```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python
from sklearn.pipeline import Pipeline

from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# 1. Feature selection
X = spaceship.drop(columns=["Transported", "PassengerId", "Name", "Cabin"])
y = spaceship["Transported"].astype(int)

numerical_columns = X.select_dtypes(include=np.number).columns.tolist()
categorical_columns = X.select_dtypes(exclude=np.number).columns.tolist()

print("Numerical columns:", numerical_columns)
print("Categorical columns:", categorical_columns)

# 2. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

# 3. Feature engineering
numerical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer([
    ("numerical", numerical_pipeline, numerical_columns),
    ("categorical", categorical_pipeline, categorical_columns)
])

print("Training shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

- Now let's use the best model we got so far in order to see how it can improve when we fine tune it's hyperparameters.

```python
from sklearn.ensemble import RandomForestClassifier

baseline_model = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=42, n_jobs=-1))
])

baseline_model.fit(X_train, y_train)
baseline_predictions = baseline_model.predict(X_test)
```

- Evaluate your model

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

def evaluate_classification_model(y_true, y_pred, model_name):
    """Print the main classification metrics for a model."""
    print(f"--- {model_name} ---")
    print(f"Accuracy:  {accuracy_score(y_true, y_pred):.4f}")
    print(f"Precision: {precision_score(y_true, y_pred):.4f}")
    print(f"Recall:    {recall_score(y_true, y_pred):.4f}")
    print(f"F1-score:  {f1_score(y_true, y_pred):.4f}")
    print("\nConfusion matrix:")
    print(confusion_matrix(y_true, y_pred))
    print("\nClassification report:")
    print(classification_report(y_true, y_pred))


evaluate_classification_model(
    y_test,
    baseline_predictions,
    "Baseline Random Forest"
)
```

**Grid/Random Search**

For this lab we will use Grid Search.

- Define hyperparameters to fine tune.

```python
# The prefix classifier__ refers to the Pipeline step named "classifier".
parameter_grid = {
    "classifier__n_estimators": [100, 200],
    "classifier__max_depth": [10, 20, None],
    "classifier__min_samples_split": [2, 5],
    "classifier__min_samples_leaf": [1, 2],
    "classifier__max_features": ["sqrt"]
}

parameter_grid
```

- Run Grid Search

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(
    estimator=baseline_model,
    param_grid=parameter_grid,
    scoring="f1",
    cv=5,
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)

print("Best hyperparameters:")
print(grid_search.best_params_)
print(f"Best mean cross-validation F1-score: {grid_search.best_score_:.4f}")
```

- Evaluate your model

```python
best_model = grid_search.best_estimator_
tuned_predictions = best_model.predict(X_test)

evaluate_classification_model(
    y_test,
    tuned_predictions,
    "Tuned Random Forest"
)

comparison = pd.DataFrame({
    "Model": ["Baseline Random Forest", "Tuned Random Forest"],
    "Accuracy": [
        accuracy_score(y_test, baseline_predictions),
        accuracy_score(y_test, tuned_predictions)
    ],
    "Precision": [
        precision_score(y_test, baseline_predictions),
        precision_score(y_test, tuned_predictions)
    ],
    "Recall": [
        recall_score(y_test, baseline_predictions),
        recall_score(y_test, tuned_predictions)
    ],
    "F1-score": [
        f1_score(y_test, baseline_predictions),
        f1_score(y_test, tuned_predictions)
    ]
})

comparison
```
